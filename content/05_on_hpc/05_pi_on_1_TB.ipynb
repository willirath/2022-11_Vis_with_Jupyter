```python
from dask import array as darr
import numpy as np
```

```python
xy = darr.random.uniform(
    size=(1 * 1_000_000_000_000 // 8 // 2, 2),  # 1 TB
    chunks=(500 * 1_000_000 // 8 // 2, 2),  # 500 MB chunks
)
r2 = (xy ** 2).sum(axis=-1)
in_circ = r2 < 1
pi = 4 * in_circ.mean()
```

```python
display(xy)
```

```python
display(pi)
```

```python
%%time

pi = pi.compute()
```

```python
pi_err = pi - np.pi
print(pi, pi_err)
```
